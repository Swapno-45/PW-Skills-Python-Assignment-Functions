## Theory Questions (Question–Answer Format)

### 1. Difference between a function and a method in Python
**Answer:** A function is a standalone block of code, while a method is a function associated with an object.

**Example:**
- Function: `len([1,2,3])`
- Method: `'hello'.upper()`

### 2. Function arguments and parameters
**Answer:** Parameters are variables in the function definition, arguments are actual values passed.

**Example:** `def add(a,b): return a+b` → `add(2,3)`

### 3. Ways to define and call a function
**Answer:** Using `def`, lambda functions, recursive functions.

**Example:** `def sq(x): return x*x`

### 4. Purpose of `return`
**Answer:** Sends result back to the caller.

**Example:** `return a+b`

### 5. Iterators vs Iterables
**Answer:** Iterable can be iterated over; iterator produces values using `__next__()`.

**Example:** List is iterable, `iter(list)` is iterator.

### 6. Generators
**Answer:** Functions using `yield` to return values lazily.

**Example:** `yield x`

### 7. Advantages of generators
**Answer:** Memory efficient, faster for large data.

### 8. Lambda function
**Answer:** Anonymous one-line function.

**Example:** `lambda x: x*x`

### 9. `map()` function
**Answer:** Applies a function to each element.

**Example:** `map(lambda x:x*2, [1,2])`

### 10. map vs reduce vs filter
**Answer:** map transforms, filter selects, reduce aggregates.

### 11. reduce() sum internal mechanism
**Answer:**
`(((47+11)+42)+13) = 113`
Step1: 47+11=58
Step2: 58+42=100
Step3: 100+13=113

## Practical Questions with Code and Output

```python
# 1. Sum of even numbers
def sum_even(lst):
    return sum(x for x in lst if x % 2 == 0)

sum_even([1,2,3,4,6])
```

```python
# 2. Reverse a string
def reverse_string(s):
    return s[::-1]

reverse_string('Python')
```

```python
# 3. Squares of numbers
def square_list(lst):
    return [x*x for x in lst]

square_list([1,2,3,4])
```

```python
# 4. Prime numbers from 1 to 200
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(1,201) if is_prime(x)]
primes
```

```python
# 5. Fibonacci iterator
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.a

list(Fibonacci(10))
```

```python
# 6. Generator: powers of 2
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

list(powers_of_two(6))
```

```python
# 7. File line by line generator
def read_file(file):
    with open(file) as f:
        for line in f:
            yield line.strip()
```

```python
# 8. Lambda sort tuples
data = [(1,3),(4,1),(2,2)]
sorted(data, key=lambda x: x[1])
```

```python
# 9. Celsius to Fahrenheit using map
celsius = [0,20,37]
fahrenheit = list(map(lambda c: (c*9/5)+32, celsius))
fahrenheit
```

```python
# 10. Remove vowels using filter
s = 'hello world'
vowels = 'aeiouAEIOU'
''.join(filter(lambda x: x not in vowels, s))
```

```python
# 11. Accounting routine using lambda and map
orders = [
    [34587, 'Learning Python', 4, 40.95],
    [98762, 'Programming Python', 5, 56.80],
    [77226, 'Head First Python', 3, 32.95],
    [88112, 'Einführung in Python3', 3, 24.99]
]

result = list(map(lambda x: (x[0], x[2]*x[3] + 10 if x[2]*x[3] < 100 else x[2]*x[3]), orders))
result
```
